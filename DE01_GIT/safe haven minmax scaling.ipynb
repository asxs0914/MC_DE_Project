```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
shv = pd.read_csv("{경로}/safe haven.csv", encoding = 'utf-8')
```

```python
shv
```

```python
shv1 = shv[['국채대비 주식수익률']]
```

```python
colnames = shv1.columns
scaler = MinMaxScaler()
#min-max값 학습
scaler.fit(shv1)

print(scaler.transform(shv1))
new_colnames = [i+'_mms' for i in colnames]

shv2 = pd.concat([shv1, pd.DataFrame(scaler.fit_transform(shv1[colnames]), columns=new_colnames)], axis=1)
```

```python
shv2
```

```python
shvq = pd.merge(left = shv, right = shv2, how = 'outer' , on = '국채대비 주식수익률')
```

```python
shvq
```

```python
shvq['국채대비 주식수익률_mms'] = shvq['국채대비 주식수익률_mms']*100
```

```python
shvq
```

```python
shvq.to_csv("safe haven_minmax2.csv", mode='a')
```
