```python
import FinanceDataReader as fdr
import pandas as pd
import numpy as np
```

```python
cpr = fdr.DataReader('KR1YT=RR') #국채
```

```python
cpr
```

```python
ks11 = fdr.DataReader('KS11', '1995-01-01')
```

```python
ks11
```

```python
cpr1 = cpr.reset_index(drop = False)
```

```python
cpr1
```

```python
cpr2 = cpr1[['Date','Close']]
```

```python
cpr2
```

```python
ks2 = ks11.reset_index(drop = False)
```

```python
ks2
```

```python
ks3 = ks2[['Date','Change']]
```

```python
ks3
```

```python
ks3['변동%'] = ks3['Change']*100
```

```python
ks3
```

```python
cnk = pd.merge(left = cpr2, right = ks3, how = 'outer' , on = 'Date')
```

```python
cnk
```

```python
cnk1 = cnk.dropna()
```

```python
cnk1
```

```python
cnk1 = cnk1[['Date', 'Close','변동%']]
```

```python
cnk1
```

```python
cnk1.columns = ['날짜','국고채권(1년)','코스피변동%']
```

```python
cnk1
```

```python
cnk1['국채대비 주식수익률'] = cnk1['코스피변동%']-cnk1['국고채권(1년)']
```

```python
cnk1
```

```python
cnk2 = cnk1.reset_index(drop = True)
```

```python
cnk2
```

```python
shv1 = cnk2[['국채대비 주식수익률']]
```

```python
shv1
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
colnames = shv1.columns
scaler = MinMaxScaler()
#min-max값 학습
scaler.fit(shv1)

print(scaler.transform(shv1))
new_colnames = [i+'_mms' for i in colnames]

shv2 = pd.concat([shv1, pd.DataFrame(scaler.fit_transform(shv1[colnames]), columns=new_colnames)], axis=1)
```

```python
shv2
```

```python
cnk3 = cnk2.copy()
```

```python
cnk3
```

```python
cnk3['국채대비 주식수익률_mms'] = shv2['국채대비 주식수익률_mms']
```

```python
cnk3
```

```python
cnk3['국채대비 주식수익률_mms'] = cnk3['국채대비 주식수익률_mms']*100
```
