```python
from selenium import webdriver
```

```python
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
def get_chrome_driver():
    # 1. 크롬 옵션 세팅
    chrome_options = webdriver.ChromeOptions()
    
    # 2. driver 생성하기
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()), # 가장 많이 바뀐 부분
        options=chrome_options
    )
    
    return driver
```

```python
driver = get_chrome_driver()
driver.maximize_window() #창이 작으면 검색버튼이 안눌리는거같아서 최대화했습니다.
```

```python
driver.get("https://www.koreagoldx.co.kr/main/html.php?agencyCode=&htmid=goods/gold_list.html")#해당 페이지로 들어감
```

```python
driver.find_element(By.XPATH,'//*[@id="contents"]/div[3]/div/div[3]/div[1]/ul[3]/li[6]/label').click() #차트에서 전체 선택
```

```python
#GS와 다르게 시군구 다 놓고 '검색'해야 결과물이 나옴 조건문으로 잘 굴려봐야함

import pandas as pd
import time
datas = pd.DataFrame()
salist=[]

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") #일단 스크롤 내림

while True:    
    try:
        for i in range(1,16):
            day = driver.find_element(By.XPATH,f'//*[@id="example-table"]/div[2]/div/div[{i}]/div[1]').text 
            buy = driver.find_element(By.XPATH,f'//*[@id="example-table"]/div[2]/div/div[{i}]/div[2]').text 
            selg = driver.find_element(By.XPATH,f'//*[@id="example-table"]/div[2]/div/div[{i}]/div[3]').text 
            sel18 = driver.find_element(By.XPATH,f'//*[@id="example-table"]/div[2]/div/div[{i}]/div[4]').text 
            sel14 = driver.find_element(By.XPATH,f'//*[@id="example-table"]/div[2]/div/div[{i}]/div[5]').text 
            golist = [[day, buy, selg, sel18, sel14]]#받은거 리스트로 저장
            dff = pd.DataFrame(golist, columns=['고시날짜','순금구매', '순금판매', '18K판매', '14K판매'])#데이터프레임으로 만들기
            datas = pd.concat([datas, dff])#빈 데이터프레임에 쌓기 
            time.sleep(1)
        print('한 페이지 완료')
        driver.find_element(By.XPATH,'//*[@id="example-table"]/div[3]/span/button[3]').click()
        
    except Exception as g:
        print('완료')
        break
        
print(datas)
```
